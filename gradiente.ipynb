```python
import torch
```

```python
def f(x):
    return (x[0] - 5)**2 + 2*(x[1] + 3)**2 # min em (5,-3)

p = torch.tensor([0.0, 0.0], requires_grad=True) # ponto inicial



```

```python
# calcular o gradiente de d no ponto p
custo = f(p)
custo.backward()
d = -(p.grad / torch.linalg.norm(p.grad)) # direcao 
print(d)
```

```python
# desvincular o ponto p e d para nao computar o gradiente
p_const = p.detach()
d_const = d.detach()
```

```python
def g(lr):
    p_novo = p_const + lr * d
    return f(p_novo)

lr = torch.tensor([0.01], requires_grad=True) # lambda
```

```python
for i in range(10):
    g_val = g(lr)
    g_derivada = torch.autograd.grad(g_val, lr, create_graph=True)[0]
    g_derivada_2 = torch.autograd.grad(g_derivada, lr)[0]

    # update lr with Newton step without tracking gradients
    with torch.no_grad():
        lr = lr - g_derivada / (g_derivada_2 + 1e-8)

    print(f"  Iteração Newton {i+1}: lambda = {lr.item():.6f}, g'(lambda) = {g_derivada.item():.6f}")

    # re-enable grad for next iteration
    lr.requires_grad_(True)

lr_optimal = lr.detach()

```
